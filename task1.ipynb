```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
data = pd.DataFrame({
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Hyderabad'],
    'Sales': [250, 300, 150, 200, 275],
    'Profit': [50, 70, 30, 45, 60],
    'Latitude': [28.6139, 19.0760, 13.0827, 22.5726, 17.3850],
    'Longitude': [77.2090, 72.8777, 80.2707, 88.3639, 78.4867]
})
print("Dataset:")
display(data)
```

```python
plt.figure(figsize=(8,5))
plt.bar(data['City'], data['Sales'])
plt.title("Sales by City (Matplotlib)")
plt.xlabel("City")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', hue='City', data=data, s=100)
plt.title("Sales vs Profit (Seaborn)")
plt.show()
```

```python
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

for i in range(len(data)):
    folium.Marker(
        location=[data.loc[i, 'Latitude'], data.loc[i, 'Longitude']],
        popup=f"{data.loc[i, 'City']} - Sales: {data.loc[i, 'Sales']}",
        tooltip=data.loc[i, 'City']
    ).add_to(india_map)

india_map
```
